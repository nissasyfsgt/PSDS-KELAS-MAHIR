##**Model Klasifikasi Naive Bayes**

Model klasifikasi Naive Bayes merupakan model klasifikasi dengan dasar gagasan Teorema Bayes dengan asumsi semua variabel prediktornya(X) adalah saling independen. Model Naive Bayes mudah untuk dibangun. 
Model ini tidak memiliki estimasi parameter iteratif yang kompleks yang membuatnya cocok untuk data yanng
sangat besar. Terlepas dari kesederhanaannya, model klasifikasi inisering kali bekerja dengan sangat baik dan banyak digunakan karena sering mengungguli metode klasifikasi yang lebih canggih. Contohnya pada klasifikasi sentimen pada dataset teks, klasifikasi email spam, dan klasifikasi dokumen.

##**Algoritma Naive Bayes**

Teorema Bayes menyajikan cara untuk menghitung probabilitas bersyarat(posterior) $P(Y=y|X=(x_1,x_2,...,x_n))$, dari $P(Y=y),P(X=(x_1,x_2,...,x_n))$ dan $P(X=(x_1,x_2,...,x_n)|Y=y)$. Model klasifikasi Naive Bayes mengasumsikan bahwa pengaruh suatu variabel prediktor pada kelas tertentu tidak bergantung pada nilai-nilai prediktor lainnya. Asumsi ini disebut sebagai class conditional independence.

Screenshot (6).png



*   $P(Y|X)$ adalah probabilitas bersyarat(posterior) dari kelas(target) dengan suatu prediktor yang
diketahui(attribut).
*   $P(Y)$ adalah probabilitas prior dari kelas.
*   $(X|Y)$ adalah likelihood yang merupakan probabilitas prediktor dengan kelas yang diketahui.
*   $P(X)$ adalah probabilitas prior dari prediktor(atribut)

##**Implementasi Pada Python**

Import Library

```python
import pandas as pd
import numpy as np
```

DataFrame Input

```python
df = pd.DataFrame({'x_1':[0,0,1,0,2,1,0,2,2,1],
                   'x_2':[0,1,2,0,2,1,2,0,1,0],
                   'y':[0,1,1,1,0,0,1,0,0,0]})
df
```

Naive Bayes untuk Data Categorical

```python
def prob_prior(df, Y):
  label_kelas = sorted(list(df[Y].unique()))
  prior = []
  for i in label_kelas:
    prior.append(len(df[df[Y]==i])/len(df))
  return prior
```

```python
def likelihood_categorical(df, nama_var, nilai_var, Y, labels):
  feat = list(df.columns)
  df = df[df[Y]==labels]
  p_x_dgsyrt_y = len(df[df[nama_var]==nilai_var]) / len(df)
  return p_x_dgsyrt_y
```

```python
def naive_bayes_categorical(df, X, Y):
  # mengambil nama kolom
  nama_kolom = list(df.columns)[:-1]
  
  # menghitung prob prior
  prior = prob_prior(df, Y)
  
  Y_pred = []
  # lopping pada setiap baris data
  for x in X:
    # menghitung likelihood
    labels = sorted(list(df[Y].unique()))
    likelihood = [1]*len(labels)
    for j in range(len(labels)):
      for i in range(len(nama_kolom)):
        likelihood[j] *= likelihood_categorical(df, nama_kolom[i], x[i], Y, labels[j])
    
    # menghitung posterior probability
    post_prob = [1]*len(labels)
    for j in range(len(labels)):
      post_prob[j] = likelihood[j] * prior[j]
    
    Y_pred.append(np.argmax(post_prob))
  return np.array(Y_pred)
  
```

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
Y_pred = naive_bayes_categorical(df, X=X, Y="y")
```

##**Implementasi Pada Python**

Import Library

```python
import pandas as pd
import numpy as np
```

DataFrame Input

```python
data = pd.read_csv('https://raw.githubusercontent.com/nissasyfsgt/DATASET/main/diabetes.csv')
data.head()
```

EDA for Distribution

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
sns.histplot(data, ax=axes[0], x="Glucose", kde=True, color='r')
sns.histplot(data, ax=axes[1], x="BloodPressure", kde=True, color='b')
sns.histplot(data, ax=axes[2], x="DiabetesPedigreeFunction", kde=True)
plt.show()
```

Naive Bayes untuk Data Numerik

```python
def prob_prior(df, Y):
  label_kelas = sorted(list(df[Y].unique()))
  prior = []
  for i in label_kelas:
    prior.append(len(df[df[Y]==i])/len(df))
  return prior
```

```python
def likelihood_gaussian(df, nama_var, nilai_var, Y, label):
  feat = list(df.columns)
  df = df[df[Y]==label]
  mean, std = df[nama_var].mean(), df[nama_var].std()
  p_x_dgsyrt_y = (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-((nilai_var-mean)**2 / (2 *std**2)))
  return p_x_dgsyrt_y
```

```python
def naive_bayes_gaussian(df, X, Y):
  # mengambil nama kolom
  nama_kolom = list(df.columns)[:-1]
  
  # menghitung prob prior
  prior = prob_prior(df, Y)
  
  Y_pred = []
  # lopping pada setiap baris data
  for x in X:
    # menghitung likelihood
    labels = sorted(list(df[Y].unique()))
    likelihood = [1]*len(labels)
    for j in range(len(labels)):
      for i in range(len(nama_kolom)):
        likelihood[j] *= likelihood_gaussian(df, nama_kolom[i], x[i], Y, labels[j])
        
    # menghitung posterior probability
    post_prob = [1]*len(labels)
    for j in range(len(labels)):
      post_prob[j] = likelihood[j] * prior[j]
    
    Y_pred.append(np.argmax(post_prob))
  return np.array(Y_pred)
```

```python
train, test = train_test_split(data, test_size=.2, random_state=41)

X_test = test.iloc[:,:-1].values
Y_test = test.iloc[:,-1].values
Y_pred = naive_bayes_gaussian(train, X=X_test, Y="Outcome")
```

```python
print(accuracy_score(Y_test, Y_pred))
```
